```python
''' IMPORTING CLASSES FROM Scikit Learn MODULE '''

from sklearn.tree import DecisionTreeRegressor as dtr
from sklearn.metrics import mean_absolute_error as mae
from sklearn.model_selection import train_test_split as tts
```

As seen, I will be using Decision Tree method

```python
features = ['Global CMP22 (vent/cor) [W/m^2]',
       'Direct sNIP [W/m^2]', 'Azimuth Angle [degrees]',
       'Tower Dry Bulb Temp [deg C]', 'Tower Wet Bulb Temp [deg C]',
       'Tower Dew Point Temp [deg C]', 'Tower RH [%]',
       'Peak Wind Speed @ 6ft [m/s]', 'Avg Wind Direction @ 6ft [deg from N]',
       'Station Pressure [mBar]', 'Precipitation (Accumulated) [mm]',
       'Moisture', 'Albedo (CMP11)', 'Cloud Type', 'Solar Zenith Angle']
X = train_data[features]
y = train_data[['Snow Depth [cm]']]
```

We have created 'X' as features list and 'y' is our targeted column

```python
train_X, test_X, train_y, test_y = tts(X, y, random_state=0)
```

This will divide the training data into two parts one for fitting (train_X and train_y) and other to check the accuracy of the model (test_X and test_y)

```python
decision_model = dtr(random_state=1)        # This will create model of Decision Tree
decision_model.fit(train_X, train_y)        # train_X and train_y are given to fit
```

```python
predict_for_error_check = decision_model.predict(test_X)
decision_error = mae(predict_for_error_check, test_y)
print(decision_error)
```

This is the mean absolute error made by the ML model when using decision tree. We can further optimize it by providing the number of splits.

** To give number of splits manually, I will use hit and trial method **

```python
def mean_abs_err(leaf, train_X, test_X, train_y, test_y):
    model = dtr(max_leaf_nodes=leaf, random_state=1)
    model.fit(train_X, train_y)
    return (mae(model.predict(test_X), test_y))
```

This function will return the mean absolute error of decision tree with the provided number of splits

```python
def get_key(inp_dict, min_ele):
    key_list = list(inp_dict.keys())
    val_list = list(inp_dict.values())
    position = val_list.index(min_ele)
    return(key_list[position])
```

This function will return the key for minimum value from the dictionary dict_of_error

```python
leaf_lst = range(100, 10000, 100)
dict_of_error = dict()
for leaf in leaf_lst:
    error = mean_abs_err(leaf, train_X, test_X, train_y, test_y)
    dict_of_error[leaf] = error
print(min(dict_of_error.values()))      # This will give the minimum error for elements in leaf_lst
leaf = get_key(dict_of_error, min(dict_of_error.values()))      #  This will extract the number of splits
print(leaf)     # This is in a set
```

Here, we can see when there are 1700 splits, the model is better fitted. So we will create new model using this number of splits.

```python
X_ = test_data[features]
new_model = dtr(max_leaf_nodes = leaf, random_state=1)
new_model.fit(train_X, train_y)
prediction = new_model.predict(X_)      # Our better prediction
```

Now we will produce the CSV file

```python
Id = np.arange(0, prediction.size)
```

```python
output = pd.DataFrame({'Id':Id , 'Snow Depth [cm]': prediction})
output.to_csv('/content/drive/MyDrive/Colab Notebooks/submitted_file.csv', index=False)      # This line will generate submission CSV file 
```
